```python
import numpy as np
import librosa as lr
from pathlib import Path
import torch
import auraloss

from kymatio_audio import scattering_audio_distance as sad

from IPython import display as ipd
plt.rcParams['figure.figsize'] = [6, 3]
```

```python
import glob

data_path = '/Users/lagrange/data/_OrchideaSOL2020_release/OrchideaSOL2020/'

classes = ['Brass', 'Strings', 'Winds']
instruments = ['Trumpet_C', 'Violin', 'Flute']
names = ['TpC', 'Vn', 'Fl']
pitch = ['C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B']

files = []
for ci, c in enumerate(classes):
    for pi, p in enumerate(pitch):
        card = data_path+c+'/'+instruments[ci]+'/ordinario/'+names[ci]+'-ord-'+p+'4-mf-*.wav'
        g = glob.glob(card)
        files.append(g[0])

audio = []
for f in files:
    data, sr = lr.load(f, sr = 22050, mono = True, duration = 1.5)   
    audio.append(data)

import scipy.io.wavfile
scipy.io.wavfile.write('demo/instruments.wav', sr, a)

a = np.hstack(audio)
ipd.Audio(a, rate=sr)
```

```python
mss_distances = sad.multiscale_audio_distance(audio, sr)
```

```python
import matplotlib.pyplot as plt
from sklearn.manifold import MDS

# from sklearn.manifold import TSNE
# emb = TSNE(n_components=2, learning_rate='auto', init='random', perplexity=3, metric='precomputed').fit_transform(distances)
# from sklearn.manifold import Isomap
# emb = Isomap(n_components=2, metric='precomputed').fit_transform(distances)

def plot_dist(distances, dist_type):

    emb = MDS(n_components=2, dissimilarity='precomputed', random_state=0).fit_transform(distances)
    h1 = plt.scatter(emb[:12, 0], emb[:12, 1], sizes=(np.arange(12)+1)*8, label='Trumpet')
    plt.plot(emb[:12, 0], emb[:12, 1], linewidth=.3)
    h2 = plt.scatter(emb[12:24, 0], emb[12:24, 1], sizes=(np.arange(12)+1)*8, label='Violin')
    plt.plot(emb[12:24, 0], emb[12:24, 1], linewidth=.3)
    h3 = plt.scatter(emb[24:, 0], emb[24:, 1], sizes=(np.arange(12)+1)*8, label='Flute')
    plt.plot(emb[24:, 0], emb[24:, 1], linewidth=.3)
    plt.legend(handles=[h1, h2, h3])
    plt.title(dist_type)
    plt.xticks([]),plt.yticks([])
    plt.savefig('demo/instrument_'+dist_type+'.png')
    plt.show()
```

```python
plot_dist(mss_distances, 'mss')
```

```python
scattering_distances = sad.scattering_audio_distance(audio, sr)
```
